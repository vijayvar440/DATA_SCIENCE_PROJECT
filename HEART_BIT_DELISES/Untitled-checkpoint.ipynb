##Introduction


This dataset appears to be a heart disease dataset (likely a processed version of the UCI Heart Disease dataset from the Cleveland database). It contains 270 patient records with 14 features, including demographic, clinical, and test results. The target variable is Heart Disease, indicating Presence or Absence of heart disease. Key columns (based on standard naming):

Age: Age in years

Sex: 1 = Male, 0 = Female

Chest pain type: 1-4 (1: typical angina, 2: atypical angina, 3: non-anginal pain, 4: asymptomatic)
BP: Resting blood pressure (mm Hg)

Cholesterol: Serum cholesterol (mg/dl)

FBS over 120: Fasting blood sugar > 120 mg/dl (1 = true, 0 = false)

EKG results: Resting electrocardiographic results (0: normal, 1: ST-T wave abnormality, 2: left ventricular hypertrophy)

Max HR: Maximum heart rate achieved

Exercise angina: Exercise-induced angina (1 = yes, 0 = no)

ST depression: ST depression induced by exercise relative to rest

Slope of ST: Slope of the peak exercise ST segment (1: upsloping, 2: flat, 3: downsloping)
Number of vessels fluro: Number of major vessels colored by fluoroscopy (0-3)

Thallium: Thallium stress test result (3: normal, 6: fixed defect, 7: reversible defect)
Heart Disease: Target - Presence or Absence

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

```python
df = pd.read_csv("Heart_Disease_Prediction.csv")
```

```python
print("data infor")
df.info()
```

```python
print ("basic Statistics : ")
df.describe()
```

```python
# TARGATE VALU DESTRIVIUTION 
```

```python
plt.figure(figsize= (8,5))
sns.countplot(x = "Heart Disease",data = df , palette = 'viridis')
plt.title  ("Destrebiution of heart Dieseqse ")
plt.xlabel("heart Disease")
plt.ylabel("count")
plt.show()
```

```python
# Numerical Features Distribution
```

```python
numerical_cols = ['Age' , 'BP' , 'Cholesterol' , 'Max HR', 'ST depression']
df[numerical_cols].hist(bins = 20 , figsize = (14,10), color = 'skyblue', edgecolor = "black")
plt.suptitle("Destribution of Numrical Features ")
plt.show()
```

```python
# Categorical Features Distribution 
```

```python
Categorical_cols = ['Sex', 'Chest pain type', 'FBS over 120', 'EKG results', 
                    'Exercise angina', 'Slope of ST', 'Number of vessels fluro', 'Thallium']  # Added closing quote here

fig, axes = plt.subplots(4,2, figsize=(15,20))
axes = axes.ravel()  

for i, col in enumerate(Categorical_cols):  # Fixed syntax by removing extra comma before i
    sns.countplot(x = col, data = df, ax = axes[i], palette='Set2')
    axes[i].set_title(f'Destribution {col}')  # Added space after 'Destribution'
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('count')

plt.tight_layout()
plt.show()
```

```python
# Age vs Heart Disease
print("Age vs Heart Disease")
plt.figure(figsize = (10,6))
sns.boxplot(x = 'Heart Disease',y = 'Age', data = df , palette='coolwarm')
plt.title('Age Distribution by Heart Disease')
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x='Age',
    hue='Heart Disease',
    multiple='stack',
    bins=20,
    palette='viridis'
)

plt.title("Age Distribution by Heart Disease")
plt.show()
```

```python
df_corr = df.copy()
df_corr['Heart Disease'] = df_corr['Heart Disease'].map({'Presence': 1, 'Absence': 0})
plt.figure(figsize= (12,10))
sns.heatmap(df_corr.corr(),
            annot = True,
            cmap = 'coolwarm',
            fmt = '.2f',
            linewidths = 0.5
)
plt.title('Correlation Matrix')
plt.show()
```

```python
# chest pain type  vs heart Disease
plt.figure(figsize=(15,10))
sns.countplot(x='Chest pain type', 
              hue='Heart Disease', 
              data=df,
              palette='muted')
plt.title("chest pain type  vs heart Desease")
plt.xlabel("chest pain type (1-4)")
plt.show()
```
